```python
import matplotlib.pyplot as plt

# Dane do porwnania
scenarios = [1, 2, 3, 4, 5]
dstar_times = [0.0535, 0.0517, 0.0483, 0.0458, 0.0499]
dstar_lite_times = [0.0465, 0.0472, 0.0483, 0.0479, 0.0480]

dstar_times_ms = [time * 1000 for time in dstar_times]
dstar_lite_times_ms = [time * 1000 for time in dstar_lite_times]

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(scenarios, dstar_times_ms, marker='o', linestyle='-', color='b', label='D*')
plt.plot(scenarios, dstar_lite_times_ms, marker='o', linestyle='-', color='r', label='D* Lite')

# Dodanie etykiet i tytuu
plt.xlabel('Scenariusz')
plt.ylabel('Czas dziaania [ms]')
plt.title('Porwnanie czasw dziaania D* i D* Lite')
plt.xticks(scenarios)
plt.grid(True)
plt.legend()

# Zapis do pliku PNG
plt.savefig('comparison_plot.png')

# Wywietlenie wykresu
plt.show()
```
